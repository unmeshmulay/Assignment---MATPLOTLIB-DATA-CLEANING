Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
# Unmesh Mulay 
# Assignment MATPLOTLIB & DATA CLEANING
# Task 1 - Program 1 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)

titanic.head()
```

```python
mfratio = list(titanic.sex.value_counts())
mfratio
```

```python
labels = 'Male', 'Female'
sizes = mfratio
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Q 2 - Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

distvals = titanic['sex'].unique()
titanic['newclm'] = np.zeros((titanic.shape[0]))
for i in range(len(distvals)):
    titanic.loc[titanic['sex'] == distvals[i], 'newclm'] = i+1
    
plt.figure(figsize=(18,5))
plt.scatter(titanic.fare, titanic.age, c=titanic.newclm)

plt.xlabel('fare', fontsize=12)
plt.ylabel('age', fontsize=12)
plt.title('Fare vs Age')
```

```python
# Unmesh Mulay 
# Assignment MATPLOTLIB & DATA CLEANING
# Task 2 - Program 1 


df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})
df
```

```python
# Q 1
# Some values in the the FlightNumber column are missing. These numbers are meant to
# increase by 10 with each row so 10055 and 10075 need to be put in place. Fill in these missing
# numbers and make the column an integer column (instead of a float column).

df.loc[df['From_To']=='MAdrid_miLAN','FlightNumber'] = 10055
df
```

```python
df.loc[df['From_To']=='Budapest_PaRis','FlightNumber'] = 10075
df
```

```python
df.loc[df['From_To']=='Budapest_PaRis','FlightNumber'] = 10075
df
```

```python
df['FlightNumber'] = df['FlightNumber'].apply(np.int64)
df
```

```python
# Q 2
# The From_To column would be better as two separate columns! Split each string on the
# underscore delimiter _ to give a new temporary DataFrame with the correct values. Assign the
# correct column names to this temporary DataFrame.

jrny = df['From_To'].str.split('_',n=1, expand = True)
jrny.rename(columns={jrny.columns[0]:'From', jrny.columns[1]:'To'}, inplace=True)
jrny
```

```python
# df['From_To'] = df['From_To'].str.capitalize()
jrny['From'] = jrny['From'].str.capitalize()
jrny['To'] = jrny['To'].str.capitalize()
jrny
```

```python
# Q 4
# Delete the From_To column from df and attach the temporary DataFrame from the previous
# questions.
df.drop(['From_To'], axis = 1, inplace=True)
df
```

```python
df['From'] = jrny['From']
df['To'] = jrny['To']
df
```
